```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(1,11)
y=np.random.rand(10)*20

plt.subplot(221)
plt.plot(x,y)

plt.subplot(222)
plt.step(x,y,color='red',linewidth=5)

plt.subplot(223)
plt.bar(x,y,facecolor="blue",edgecolor='black')

plt.subplot(224)
plt.barh(x,y,facecolor="gray", edgecolor='blue')

plt.show()
```

# El paquete matplotlib dispone de gran variedad de graficos:

```python
import numpy as np
import matplotlib.pyplot as plt

plt.subplots(2,3,figsize=(20,10))

x1=np.arange(1,6)
y1=np.array([[1,1,2,3,7], [0,4,2,6,8], [1,3,5,7,9]])
plt.subplot(231)
plt.stackplot(x1,y1,labels=["Abril ","Mayo","Junio"])
plt.xlabel("Diagrama de pila (stackplot)")
plt.legend(loc='upper left')

x2=np.linspace(0.1,2*np.pi,41)
y2=np.exp(np.sin(x2))
plt.subplot(232)
plt.stem(x2,y2,use_line_collection=True)
plt.xlabel("Grafico de tallo(stem plot)")

x3=np.arange(0.1,4,0.5)
y3=np.exp(-x3)
plt.subplot(233)
plt.errorbar(x3, y3, xerr=0.2,yerr=0.4)
plt.xlabel("Barras de error (errorbar)")

plt.subplot(234)
plt.text(0.6,0.7,"python",size=40,rotation=30,ha="center",va="center",
         bbox=dict(boxstyle="round",facecolor='yellow'))
plt.text(0.7,0.6,"matplotlib",size=40,rotation=-25,ha="right",va="top",
         color='blue',bbox=dict(boxstyle="square",edgecolor='red',
                                facecolor='lightgray'))
plt.xlabel("Textos en etiquetas")

x4=np.arange(1,11)
y4=np.random.rand(10)*20
plt.subplot(235)
plt.plot(x4,y4)
plt.annotate('importante',xy=(x4[2],y4[2]),xytext=(x4[2]+3, y4[2]+3),
             arrowprops=dict(facecolor='gray'))
plt.xlabel("Anotaciones en un grafico (annotate)")

plt.subplot(236, projection='polar')
plt.plot(x4,y4)
plt.xlabel("Proyeccion polar de grafico de lines")

nombre=input("dime el nombre de los archivos:")
plt.savefig(nombre + ".pdf", facecolor="Salmon", transparent=True, bbox_inches="tight",
                         metadata=dict({"Creator":"Python y matplotlib",
                                        "Author":"Alumno cepibase",
                                        "title":"Graficos de ejemplo"}))
print("Archivo generado")
plt.savefig(nombre+".svg",facecolor="yellow")
print("Archivo svg generado")
plt.savefig(nombre, facecolor="LightBlue", dpi=45)
print("Archivo png generado")

plt.show()
```
