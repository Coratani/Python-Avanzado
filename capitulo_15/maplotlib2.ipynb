# Algunos tipos de graficos de matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(1,11)
y=np.random.rand(10)*20

plt.plot(x,y)
plt.show()
```

## Grafico de escalon(step)

```python
plt.step(x,y, color='red', linewidth=5)
plt.show()
```

## Grafico de barras(bar)

```python
plt.bar(x,y, facecolor="blue", edgecolor='black')
plt.show()
```

## Grafico de barras horizontales(barh)

```python
plt.barh(x,y,facecolor="gray",edgecolor='blue')
plt.show()
```

## Grafico de sectore(pie)

```python
lista = np.array([['Francia','Espaa','EEUU','China','Italia'],
                 [86.9,81.8,75.9,60.7,58.2]])
explode=[0,0.2,0,0.5,0]#Destacar espa y China
plt.pie(lista[1],labels=lista[0],explode=explode,
       autopct='%1.1f%%',shadow=True,startangle=90,rotatelabels=30)
plt.title("TOP DESTINOS TURISTICOS EN 2020",pad=20)
plt.show()
```
