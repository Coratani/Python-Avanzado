### Primeros pasos con MATplotlib

```python
import matplotlib.pyplot as plt

meses =["Enero","Febrero", "Marzo","Abril"]
pedro =[1550, 1830, 1700, 1860]
ana=[1420, 1650,1720,1930]
plt.plot(meses, pedro, label="Gastos Pedro")
plt.plot(meses,ana, label="Gastos Ana")
plt.xlabel("Prime cuatrimestre")
plt.ylabel("Cantidades â‚¬")
plt.legend()
plt.show()
```

### Varias series se pueden unificar en una misma funcion plot

```python
plt.plot(meses, pedro,"-h",meses, ana, ":or")
plt.grid(axis="y", linewidth="2")
plt.legend(['Gastos Pedro','Gastos Ana'])
plt.show()
```

### Integracion con NumPy

```python
import numpy as np
datos = np.random.rand(5,3)*500
datos = np.sort(datos,axis=1)
titulos=np.array(["uno","dos","tres","cuatro","cinco"])
print(datos)
for fila in zip(datos, titulos):
    plt.plot(fila[0], fila[0]**2,label=fila[1])
plt.legend()
plt.show()
```

### Graficos trigonometricos

```python
tiempo = np.arange(0.1,20,0.1)
y1=np.sin(tiempo)/tiempo
y2=np.sin(tiempo)*np.exp(-tiempo)
plt.plot(tiempo, y1, tiempo, y2)

plt.text(2,0.5,r'$\frac{\sin(x)}{x}$}', fontsize=20)
plt.text(13, 0.2,r'$\sin(x) \cdot e^{-x}$', fontsize=16)

plt.title('Representacion de dos funciones')
plt.xlabel('Tiempo / s')
plt.ylabel('Amplitud / cm')
plt.grid()
plt.show()
```
