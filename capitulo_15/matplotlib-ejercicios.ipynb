##### Realice la importación de NumPy y el módulo pyplot de matplotlib, para utilizarlos en los ejercicios.
##### Cree un array con NumPy llamado capitales, con las 5 siguientes capitales: Londres, París, Roma, Rabat y Atenas, que representarán lugares donde se han producido ventas de varios sectores, y formarán el eje x.
##### Cree otro array llamado autos, con 5 valores aleatorios entre 1500 y 3000 (serán ventas en automoción en millones de €).
##### Cree otro array llamado construc, con 5 valores aleatorios entre 3200 y 4500 (serán ventas en el sector de la construcción, en millones de €).
##### Cree un último array llamado turis, con 5 valores aleatorios entre 2500 y 4000 (serán ganancias en el sector turístico, en millones de €).
##### Los valores de los tres arrays deberán tener 3 decimales. Imprima el contenido de todos ellos para entender mejor los valores con los que trabajarán los gráficos siguientes.

```python
import numpy as np
import matplotlib.pyplot as plt

capitales = np.array(["Londres", "París", "Roma", "Rabat", "Atenas"])
autos = np.around(1500+np.random.rand(5)*(3000-1500), decimals=3)
construc = np.around(3200+np.random.rand(5)*(4500-3200), decimals=3)
turis = np.around(2500+np.random.rand(5)*(4000-2500), decimals=3)
print("Capitales:\n", capitales, "\nAutomoción:\n", autos, "\nConstrucción:\n", construc, "\nTurismo:\n", turis)
```

##### Muestre un gráfico de líneas con la progresión de las ventas y ganancias generadas anteriormente, en las capitales indicadas. Deberá tener las siguientes características:
##### Serie de automoción: Marca de triángulo invertido (triangle_down), línea normal (solid line) y color verde (green). Leyenda: "Automoción"
##### Serie de construcción: Marca de diamante (diamond), línea discontinua (dashed line) y color azul (blue). Leyenda: "Construcción"
##### Serie de turismo: Marca de estrella (star), línea punteada (dotted line) y color amarillo (yellow). Leyenda: "Turismo"
##### Etiqueta eje X: "Principales capitales", eje Y: "Cantidades millones €"
##### Mostrar leyenda, su posición (loc) será "best". Mostrar también cuadrícula.

```python
plt.plot(capitales,autos,"-vg",capitales,construc,"--Db",capitales,turis,":*y" )
plt.xlabel("Principales Capitales")
plt.ylabel("Cantidades Millones €")
plt.legend(['Automocion','Construccion','Turismo'],loc="best")
plt.show()
```

##### Muestre un gráfico de sectores para el sector de automoción en las capitales indicadas; dicho gráfico deberá mostrar el porcentaje de los valores con 1 decimal, mostrar sombra, comenzar en el ángulo 180 y las etiquetas de las capitales deben rotarse 25 grados.
##### Debe aparecer separada la cuña con el valor más grande; para ello, debe crear inicialmente un array con separación 0 para todas las cuñas, después averiguar la posición del valor máximo en automoción, y finalmente en el array de las separaciones, aplicar el valor de 0.2 para la posición averiguada.
##### El gráfico deberá tener el título "SECTOR DE AUTOMOCIÓN" con una separación (pad) de 30.

```python
explode = np.zeros(5) 
maximo = np.argmax(autos)
explode[maximo] = 0.2

plt.pie(autos, labels=capitales, explode=explode,autopct='%1.1f%%', shadow=True, startangle=180, rotatelabels=25)
plt.title("SECTOR DE AUTOMOCIÓN", pad=30)
plt.show()
```

##### Muestre una agrupación de 3 gráficos de sectores, similares al realizado en el ejercicio anterior, pero para los tres sectores iniciales (automoción, construcción y turismo). En cada uno de ellos debe destacarse la cuña con valor más alto.
##### Inicialmente establezca una cuadrícula de 1 fila y 3 columnas, asignando a cada gráfico un tamaño de 20 x 10.

```python
plt.subplots(1, 3, figsize = (20, 10))
explode = np.zeros(5)
maximo = np.argmax(autos)
explode[maximo] = 0.2 

plt.subplot(131)
plt.pie(autos, labels=capitales, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle=180, rotatelabels=25)
plt.title("SECTOR DE AUTOMOCIÓN", pad=30)
explode = np.zeros(5)
maximo = np.argmax(construc)
explode[maximo] = 0.2 

plt.subplot(132)
plt.pie(autos, labels=capitales, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle=180, rotatelabels=25)
plt.title("SECTOR DE CONSTRUCCIÓN", pad=30)
explode = np.zeros(5) 
maximo = np.argmax(turis)
explode[maximo] = 0.2

plt.subplot(133)
plt.pie(autos, labels=capitales, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle=180, rotatelabels=25)
plt.title("SECTOR DE TURISMO", pad=30)
plt.show()
```
